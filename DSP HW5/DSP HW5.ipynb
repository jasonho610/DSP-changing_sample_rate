```python
samplerate, data = wavfile.read("input.wav")
samplerate
```

```python
length_time = len(data) / samplerate
length_time
```

```python
time = np.linspace(0., length_time, len(data))
len(data)
```

```python
data_left = []; data_right = []
for i in range(len(data)):
    data_left.append(data[i][0])
    data_right.append(data[i][1])
```

```python
pd.DataFrame(data_left[-11:-1])
```

```python
pd.DataFrame(data_right[-11:-1])
```

```python
plt.figure(figsize = (20,5))
plt.plot(time, data_left, label = "data", color = "blue")
plt.title("data in wav file")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend(loc = 1)
plt.show()
```

```python
plt.figure(figsize = (20,5))
plt.plot(time, data_right, label = "data", color = "orange")
plt.title("data in wav file")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend(loc = 1)
plt.show()
```

```python
plt.figure(figsize = (20,5))
plt.plot(time, data_left, label = "data", color = "blue")
plt.plot(time, data_right, label = "data", color = "orange")
plt.title("data in wav file")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend(loc = 1)
plt.show()
```

```python
x = data_left # get the left channel data
L = 80 # up sampling  4*16000 = 64000
M = 441 # down sampling 64000/8 = 8000
wc=np.minimum(np.pi/L, np.pi/M) # low pass fir filter cutoff frequency
B=ss.firwin(numtaps=17, cutoff=wc, window='hamming') # FIR low pass filter coefficients (17 taps)
xu = np.zeros(L*len(x))
xu[0:-1:L] = x # up sampling from x
yu = ss.lfilter(B,[1],xu) # pass FIR filter with hamming window
yd = yu[0:-1:M] # down sampling from yu
```

```python
downsampling_time = np.linspace(0., length_time, len(yd))
print(length_time, len(yd), len(x), len(yd)/length_time)
```

```python
plt.figure(figsize = (20,5))
plt.plot(downsampling_time, yd, color = "blue", label = "down-sampling")
plt.title("Down Sampling")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend(loc = 1)
plt.show()
```
